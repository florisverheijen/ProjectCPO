ik wga een PCA van de data set Movies maken. 

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as la
import pandas as pd

#import movies data set
movies=pd.read_csv("C:\\Users\\s144239\\Documents\\GitHub\\ProjectCPO\\pandas\\Movies.csv")
m=movies.head() # this shows the first 5 rows of the DataFrame
print(m)
```

```python
#lengte van kolommen 
n=len(movies.columns)
print(n)
```

```python
#alleen de kolommen 2 tot len(movies) wil ik een dataframe maken
m=movies.loc[:,movies.columns[2:n-1]]
print(m.head())
```

```python
from sklearn.preprocessing import StandardScaler

#this is the function for normalizing 
X = StandardScaler().fit_transform(m)
#print(X)

normalised_movies = pd.DataFrame(X,columns=movies.columns[2:n-1])
print(normalised_movies.head())
```

```python
#berekent de cov matrix
covm=np.cov(X.T)
```

```python
#calculate the eigenvalues and eigenvectors
paireig=la.eig(covm)
#make a list with tulpe of eigenvalues and eigenvectors
eig_pairs = [(np.abs(paireig[0][i]), paireig[1][:,i]) for i in range(len(paireig[0]))]


eig_pairs.sort 
eig_pairs.reverse 
#print(eig_pairs)

#print the eigenvalues in the correct sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
#calculate total variance by suming the diagonal of the cov-matrix
totalevariance=np.trace(covm)
#calculate each principal component
componentmatrix= [(i / totalevariance)*100 for i in sorted(paireig[0], reverse=True)]
#print(componentmatrix)
cum_var_exp = np.cumsum(componentmatrix) # python
```

```python
#make a bar plot of the principal components
plt.bar(range(len(componentmatrix)),componentmatrix)
plt.step(range(len(componentmatrix)),cum_var_exp)
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
plt.show()
```

```python
matrix_w = np.hstack((eig_pairs[0][1].reshape(35,1),
                      eig_pairs[1][1].reshape(35,1)))

print('Matrix W:\n', matrix_w)
```

```python
#do a inner product with the data
Y = X.dot(matrix_w)
#print (Y)
#print(len(Y))
```

```python
#make a plot of the data
from matplotlib import pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Romance in films",fontsize=20)
targets = [1, 0]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = movies['Romance'] == target
    plt.scatter(Y[indicesToKeep,0]
               , Y[indicesToKeep,1], c = color, s = 50)

plt.legend(('romance films', 'no romance films'),prop={'size': 15})
plt.show()
```
