#### Step 2: Calculation of the covariance matrix

4) Calculate the Covariance-matrix using the command *cov*

**Answer:**

```python
cov_mat = np.cov(X.T) 
cov_mat
#print('NumPy covariance matrix: \n%s' %np.cov(X.T))
```

#### Step 3 Calculation of the Eigenvalues and Eigenvectoren

5) Calculate the Eigenvalues and Eigenvectoren<br>
**Answer:**<br>
with *eig_vals, eig_vecs = np.linalg.eig(cov_mat)*

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

eig_vecs
```

#### Step 4: Determining the most important principal componentÂ Â 

6) make a eig_pairs list of (eigenvalue, eigenvector) tulples and print the Eigenvalues in descending order.<br>
**Short answer:**<br>
with make a list of tulples<br>
.sort<br>
.reverse<br>
print Eigenvalues in descending order<br>
**Long answer:**<br>

```python
#selecting principal components


# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
#eig_pairs.sort(key=lambda x: x[0], reverse=True)
# or
eig_pairs.sort() # python
eig_pairs.reverse() # python


# Visually confirm that the list is correctly sorted by decreasing eigenvalues
#print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

#### Step 5: Calculation of the explained variance

7.1) Calculate the totale varaince of the eigenvalues.<br>
**Answer:**

```python
tot = sum(eig_vals)
```

7.2) Calculate the explained variance (var_exp) <br>
**Answer:**

```python
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)] # percentage berekenen van elke PC
```

7.3) Calculate the Cummulative explained variance (cum_var_exp)<br>
**Answer:**

```python
cum_var_exp = np.cumsum(var_exp)
```

#### Step 6 Cummulative plot of the principal components.

8) Make in one plot a bar graph of the individual explained variance and a step graph of the cumulative explained variance.<br>
**Answer:**

```python
plt.figure(figsize=(6, 4))
plt.bar(range(30), var_exp, alpha=0.5, align='center',
         label='individual explained variance')
plt.step(range(30), cum_var_exp, where='mid',
         label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

9) Make a scree plot. <br>
**Answer**

```python
fig = plt.figure(figsize=(8,5))
plt.plot(var_exp, 'o-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
 
leg = plt.legend(['Eigenvalues'])
plt.show()
```

#### Step 7: Reducing the dimensions


10) Reshape the eig_pair data in matrix_w with only the Eigenvectors of the two highest eigenvalues. *hint: use hstack*<br>
**Answer:**

```python
 
matrix_w = np.hstack((eig_pairs[0][1].reshape(30,1),
                      eig_pairs[1][1].reshape(30,1)))

print('Matrix W:\n', matrix_w)
```

11) calculate the inner product of X and matrix_w and name it **Y**<br>
**Answer:**

```python
Y = X.dot(matrix_w)
```

#### Step 8: Plotting the data 

12) Make a scatter plot that display the data of bennign in red and maligant in green with the plottitle *Principal Component Analysis of Breast Cancer Dataset*,<br> x axis Principal Component 1, y axis Principal Component 2 and legend.<br>
**Answer:**

```python
from matplotlib import pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(Y[indicesToKeep,0]
               , Y[indicesToKeep,1], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
plt.show()
```
