# 1. Function Calls and Libraries

## Learning objectives

After this section, you should

* Know how to use **functions** to work with data.
* Know how to **import libraries**.

## Introduction

Expressions can also contain **function calls**
that execute code defined elsewhere.
These functions are either built-in, taken from a library,
or defined in your program.

## Functions for code reuse

For many operations on data there exist predefined functions.
You don't have to write the code yourself,
but instead can reuse code developed by others.
We will do that a lot in these lessons.

We distinguish three kinds of predefined functions:

* **built-in** functions, which can be used directly;
* functions from [**The Python Standard Library**](https://docs.python.org/3/library/)
    that are available in any Python installation but you need to **import** them first;
* functions from **third-party libraries**,
    which need to be installed separately and also need to be imported.

## Function calls

Using a function is also known as a **function call**.
In general, the syntax of a function call of the function named `f` is:

```python
f(argument_1, argument_2, ...)
```

Notes:

* Some functions are part of an object.
    Use the syntax `obj.f` to call the function `f` of object `obj`.  
    Such a function call then also includes the object as an extra argument.
* For a function `f` from a library named `lib`, you write `lib.f` as its full name.

The arguments in the argument list can have various syntactic forms.
The two most common forms are:

* `expression` (also known as **positional argument**)
* `name=expression` (also known as **keyword argument**)

Positional arguments must appear _before_ keyword arguments.
Thus,
a function call of the form `f(name=expression, expression)` is invalid.

The semantics of a function call are:

1. evaluate the expressions in the arguments and
    bind them to the corresponding function **parameters**  
    (correspondence is based on position for positional arguments,
    and on the included `name` for keyword arguments);  
1. bind any function parameters not bound at this point
    to their **default** value  
    (if the function does not define a default value for an unbound parameter,
    then an error is raised);
1. transfer control to the code of the function;
1. when execution of the function's code terminates,
    control is transferred back to the caller with a **result**
    (which could be `None`);
1. the returned result is the value of the function call.

### Documentation of functions

Recall that you can obtain documentation about function `f` in two ways:

* execute a code cell containing `f?`, or
* type Shift-Tab or Shift-Tab-Tab with the cursor after `f`.

**Notes** about reading documentation for functions:

* The function documentation mentions the function's _name_,
    its _parameters_, and _default parameter values_:
```python
    f(parameter_def_1, parameter_def_2, ...)
```
    where each parameter definition can take one of two forms:
    
    * `name` : a parameter _without default value_ (the value _must_ be supplied in the function call)
    * `name=expression` : a parameter **with default value** (another value _may_ be supplied in the function)

* Nowadays, parameter definitions often have type information:

    * `name: type`
    * `name=expresssion: type`
    
* The documentation also states assumptions about argument values,
    and how the function effect or result relates to those arguments.

## Built-in functions

Some functions are needed so often that they are _built in_.
We divide them into groups based on their purpose:

* type conversions
* numeric operations
* sequence operations, applicable to strings and lists
* string operations
* list operations, especially list modifications
* input-output functions

We mention the most relevant functions in each group below.

For details see
_The Python Standard Library_:
* [Built-in functions](https://docs.python.org/3/library/functions.html)
* [Sequence Types](https://docs.python.org/3/library/stdtypes.html#sequence-types-list-tuple-range)
* [Text Sequence Type](https://docs.python.org/3/library/stdtypes.html#text-sequence-type-str)

## Type conversions

* `int(x)` : convert number or string `x` to `int`; `int()` returns `0`; does _not_ round floating-point numbers
* `float(x)` : convert number or string `x` to `float`; `float()` returns `0.0`
* `bool(x)` : convert `x` to `bool`; `bool()` returns `False`; any _empty_ value converts to `False`
* `str(x)` : convert `x` to `str`; `str()` returns the empty string
* `list(x)` : convert sequence `x` to `list`; `list()` returns the empty list

## Numeric operations

* `abs(x)` : return absolute value of number `x`
* `max(x, y, ...)` : return maximum of `x`, `y`, ...; also works on strings and lists using lexicographic order
* `min(x, y, ...)` : return minimum of `x`, `y`, ...; also works on strings and lists using lexicographic order
* `round(x)` : return number `x` rounded to `int`; **rounds half to even**!

## Sequence operations

* `len(x)` : return length of sequence `x`
* `max(x)` : return largest item in sequence `x`
* `min(x)` : return smallest item in sequence `x`
* `sorted(x)` : return sorted list of items in `x` (in ascending order)
* `sorted(x, reverse=True)` : return reverse sorted list of items in `x` (in descending order)
* `x.count(y)` : return number of times that `y` occurs as item in sequence `x`

## String operations

* `x.join(list_of_strings)` : return the string obtained by inserting `x` between adjacent items in the list
* `x.lower()` : return `x` with all letters converted to lower case
* `x.split(y)` : return list of strings obtained from string `x` by cutting it at every occurrence of `y`;
    if `y` is omitted, `x` is cut at whitespace
* `x.strip()` : return `x` after removing all leading and trailing whitespace
* `x.upper()` : return `x` with all letters converted to upper case

Examples:

````python
','.join(list('abc'))  # 'a,b,c'
````

````python
' A b\n c\n'.lower()  # ' a b\n c\n'
````

````python
'a,b,c'.split(',')  # ['a', 'b', 'c']
````

````python
' A b\n c\n'.split()  # ['A', 'b', 'c']
````

````python
' A b\n c\n'.strip()  # 'A b\n c'
````

````python
' A b\n c\n'.upper()  # ' A B\n C\n'
````

## List modifications

* `x.append(y)` : append item `y` to list `x`; same as `x.extend([y])`
* `x.extend(y)` : extend list `x` with all items from list `y`; same as `x += y`
* `x.insert(i, y)` : insert item `y` in list `x` at index `i`;
    the items at index `i` and beyond are first moved up one position
* `x.pop(i)` : remove and return the item at index `i` from list `x`;
    if `i` is omitted, it is taken as `-1`
* `x.reverse()` : reverse the order of the items in list `x`

## Input-output functions

* `print(x, ...)` : print values `x`, ... to standard output;
    values are separated by `sep`, and the last one is followed by `end`;
    default values for these keyword parameters are: `sep=' ', end='\n'`
    
* `open(x)` : open file at path `x` (a string) for reading;
    always use this function in a `with` statement
```python
with open(x) as f:
    lines = f.readlines()
```

* `open(x, 'w')` : open file at path `x` (a string) for writing;
    
    **NOTE**: **this will delete any previous content** if the file already existed;

    always use this function in a `with` statement
```python
with open(x, 'w') as f:
    print(y, ..., file=f)
```

# 2. Aliasing and Sharing

## Learning objectives

After this section, you should

* Know about **aliasing** and **sharing**.
* Know how to deal with **aliasing** and **sharing**.

## Introduction

The main reason for having mutable objects in Python is
to improve the **efficiency** of programs.
Without mutable lists,
changing a single item in a list would require constructing a new list object
by copying over all items except the one that is changed.
Mutable objects can save on _execution time_ and _memory usage_.

However,
this improvement in efficiency comes at a price.
The network of bindings can become complex,
and it is easy to get lost in it.

Consider three lists `x`, `y`, and `z`,
that all get assigned the same value but in different ways:

````python
x, y = ['a', 'b', 'c'], ['a', 'b', 'c']
z = y
x, y, z
````

Variables `x` and `y` refer to _different_ objects,
whereas `y` and `z` refer to the _same_ object.
Variables `y` and `z` are **aliases**:
different names for the same object.

We can check this by obtaining their identities:

````python
id(x), id(y), id(z)
````

Now, let's modify the value of `y`,
and check that the variables still refer to the exact same objects as before the assignment.

````python
y[1] = 4
id(x), id(y), id(z)
````

Finally, we inspect the values of these variables:

````python
x, y, z
````

Observe that both `y` and `z` turn out to be changed,
even though we did not assign a new value to `z[1]`.
We also say that `y` and `z` **share** their data.
This saves memory,
but also links their fates:
if you touch one, you touch the other.

Note that also list items bind objects, just like variables.
Thus, sharing can be less obvious.
Another example with three lists `u`, `v`, and `w`:

````python
u = [3, 1, 4, 1, 5, 9]
v, w = ['v', u], ['w', u]
u, v, w
````

And now we modify `w` and inspect all variables again:

````python
w[1][2] = 'X'
u, v, w
````

All three lists share the object referred to by `u`.

## Copying to avoid sharing

If you really want to be able to change `u`, `v`, and `w` independetly,
then you will need to make a **copy** of `u` when putting it in `v` and `w`:

````python
u = [3, 1, 4, 1, 5, 9]
v, w = ['v', u[:]], ['w', u[:]]
u, v, w
````

Now modify `w` and inspect all variables:

````python
w[1][2] = 'X'
u, v, w
````

Only `w` changed.
The price is that this costs more memory,
because multiple copies are stored.

So, as a Python programmer you have to make this trade-off.

## Data frames

The most important data type of the Pandas library is **`pd.DataFrame`**.
It is a _composite_ data type, whose values are called **data frames**.

A data frame is a two-dimensional arrangement of data values,
typically of a primitive data type (`int`, `float`, `bool`, `str`,
but also some variants of these as offered by the NumPy library;
details are not relevant at this point).

Here is an example of a data frame; we name it `df`.
It has

* four rows indexed from 0 through 3, and
* three columns labeled `'A'`, `'B'`, and `'C'`.

````python
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': ['one', 'two', 'three', 'four'], 'C': [False, True, False, True]})
df
````

The following diagram illustrates some terminology:



* A **row** is a horizontal selection of data values.
* A **column** is a vertical selection of data values.
* The **index** provides an identification of the rows.
* The **column labels** provide an identification of the columns.

Typically, all values in one column have the same type.
But values in a row can be of different types.

### `Index` and `Series` objects

To understand Pandas data frames better,
you should know about two more data types:

* **`pd.Index`**:
    this is a special type of object that holds the index;
    by default, it is just an interval of integers starting from 0.
* **`pd.Series`**:
    this type represents a single named sequence of data values of the same type,
    with an index.  
    It is used for a single column in a data frame,
    but also for a row extracted from a data frame,
    where the column labels then serve as the `Series` index,
    and the row index serves as the `Series` name.

Thus, a data frame is a sequence of `Series` objects,
with a shared `Index` object.
The names of the `Series` objects are the column labels.


## Information about a `DataFrame`

The following attributes provide information about the `DataFrame df`.

* **`df.shape`** : the number of rows and number of columns
* **`df.index`** : the row index
* **`df.columns`** : the column labels

````python
df.shape
````

````python
df.index
````

````python
df.columns
````

The `DataFrame` function **`info()`** returns
an overview of all information about a data frame:

````python
df.info()
````

## Getting data from a `DataFrame`

You can **get a column** from a data frame by indexing, just like a sequence.
The result is a `Series` object.
This is useful to extract a particular feature from the data set.

````python
df['B']
````

To **get a row** from a data frame,
you use the `DataFrame` attribute `loc`.
It is called an _indexer attribute_,
because it supports indexing
with _square brackets_, just like indexing of lists.
The result is also a `Series` object:

````python
df.loc[2]
````

You can **get a value** at a given location in a data frame
in several ways:

* **`df[column_label][row_index]`** : first get the column,
    then get the value from the resulting `Series` object
* **`df.loc[row_index, column_label]`** : get the value directly,
    using `loc` with the row index and column label

````python
df['B'][2]
````

````python
df.loc[2, 'B']
````

Slicing can be used to **get larger parts from a data frame**,
with a syntax similar to the slicing of lists.

> NOTE: When slicing `DataFrame` and `Series` objects
> with the syntax `.loc[start:stop]`,
> the **`stop` value is included**.  
> This contrasts to _list_ slicing, where `stop` is _not_ included.

To select a slice of rows,
use `.loc[start:stop]`.
The result is a (new!) data frame:

````python
df.loc[1:2]
````

The `DataFrame` function call **`df.head(n)`** returns the first `n` rows of the data frame `df`,
regardless of how the rows are indexed.
The default value for `n` is 5.

````python
df.head(2)
````

You can also apply slicing to the columns:

````python
df.loc[1:3, 'A':'B']
````

When `start` or `stop` is omitted in the slice `start:stop`,
the first or last item is implied.  
Therefore, `:` extracts all rows or columns:

````python
df.loc[:, 'A':'B']
````

To **get _non-adjacent_ rows**,
use `.loc` with a _list of row indexes_
(note the **double square brackets**).
The result is a (new!) data frame:

````python
df.loc[[1, 3]]
````

You can also **get _non-adjacent_ columns**.
One way is by providing a _list of column labels_ directly to the data frame
(note the **double square brackets**).
The result is a (new!) data frame:

````python
df[['A', 'C']]
````

An alternative way of getting non-adjacent columns,
is to use `.loc` with a _full slice_ for the rows using just '`:`':

````python
df.loc[:, ['A', 'C']]
````

Note that `df['B']` returns column '`B`' as a `Series` object.
Sometimes it is more convenient to get it as a `DataFrame` with one column.
This is achieved by using a list with only `'B'` as column label
(note the **double square brackets**):

````python
df[['B']]
````

## Reading data from file into a `DataFrame`

It is a good habit to look at the file contents first,
before reading it in.
We can do that right in the notebook.

In these lessons,
we will mainly use the **comma-separated values** (**CSV**) format
for storing simple data in a file.

Our CSV files will (usually) have

* one header line with column labels,
* each row on a line of its own,
* commas to separate the fields in each row
    (in some cases, you may encounter semicolons or TABs),
* strings enclosed in double quotes
    (in some case, strings are not enclosed), and
* floating-point numbers using a decimal point (not a comma).

The example file is named `country-data.csv`.
It was adapted from this repository: https://github.com/lorey/list-of-countries.

This file is located in the `datasets` directory.

````python
datadir = '../../datasets/country-data/'
file_countries = datadir + 'country-data.csv'
````

To open the file for reading and show the first five lines,
we use the following recipe:

````python
with open(file_countries) as f:
    lines = f.readlines()[:5]
lines
````

Apparently,
this CSV file uses commas as field separators,
and strings are enclosed by double quotes
(thus, strings can contain commas).

We use the function **`pd.read_csv`** to read it into a `DataFrame`
that we will name `country_data`.

We do not have to supply the argument **`sep=...`** to set the field separator,
because its default value is a comma.

We do need the arguments

```python
na_values=[''], keep_default_na=False
```

becaues '`NA`' is used to encode the continent _North America_
(see the line shown above with `"Antigua and Barbuda"`),
and by default this string is interpreted as _Not Available_.

It is a good habit to inspect the head of that data frame
after reading it in:

````python
country_data = pd.read_csv(file_countries, na_values=[''], keep_default_na=False)
country_data.head()
````

Let's have a look at the information about this data frame:

````python
country_data.info()
````

We see that there are 251 rows.
Therefore, some columns have missing values:

* `tld` misses 2 values
* `capital` misses 6 values

These are the seven columns of data and their types.

* `name` (string): country name in English
* `alpha_3` (string): 3-letter ISO 3166-1 country code
* `tld` (string): top-level domain in Domain Name System
* `continent` (string): 2-letter continent code
* `capital` (string): capital city in English
* `area` (float): area in square kilometer
* `population` (integer): population (measurement date/source unclear; cf. https://geonames.org/)

## Summary statistics

**Summary statistics for _numeric_ data** in a data frame `df`
can be obtained by **`df.describe()`**:

````python
country_data.describe()
````

To get summary statistics for **all columns** supply the argument **`include='all'`**:

````python
country_data.describe(include='all')
````

> NOTE: **NaN** represents a **missing value**,
> also known as **not available** or **null value**.

From these summary statistics for the country data, we see that

* names and alpha-3 codes are unique (maximum frequency equals 1)
* 3 countries share the top-level domain `.gp`
* there are 7 continents, and Africa has the most countries (58)
* 2 capitals are named Kingston, and there are no other such duplicates

### Separate statistics

There are also functions to obtain separate statistics for `DataFrame df`:

* **`df.count()`** : Return the **number** of values
* **`df.sum()`** : Return the **sum** of the values
* **`df.mean()`** : Return the **mean** of the values
* **`df.std()`** : Return the **sample standard deviation** of the numeric values,
    normalized by $N-1$
* **`df.min()`** : Return the **minimum** of the values
* **`df.max()`** : Return the **maximum** of the values
* **`df.median()`** : Return the **median** of the values
* **`df.quantile(q)`** : Return the value at the given **percentile** `q`;
    `q=0.5` is default, giving the **median**

Notes:

* **NA/null values** are skipped by default.
* If the data frame has more than one column with relevant data,
    then a `Series` object is returned with the statistic for each column.

Let's compute the total _area_ and _population_:

````python
country_data[['area', 'population']].sum()
````

## Sorting a `DataFrame`

To sort a data frame `df` by the values in column `'c'` use **`df.sort_values(by='c')`**.
Note that this does not modify the data frame,
but rather returns a new data frame.  
A series is sorted by the same function,
but it does not need the `by` argument.

Here is the country data sorted by _population_:

````python
country_data.sort_values(by='population')
````

To sort in **descending order**, supply the argument **`ascending=False`**:

````python
country_data.sort_values(by='area', ascending=False)
````

> A data frame can be **sorted in place**,
> that is, the data frame is modified,
> rather than returning a new data frame.
> This is achieved by supplying the argument **`inplace=True`**,
> but is not recommended (rather, assign the result to a new variable).

To **sort on multiple columns**, provide a _list of column labels_:

````python
country_data.sort_values(by=['continent', 'name'])
````

As you can see, the index (left-most column) labels are sorted along as well.
If you want to have a new index for the result, in sorted order,
then you can use the function **`reset_index()`**.
The old index is preserved as well in the column labeled `'index'`.

Let's sort the country data by _population_,
reset the index, and assign the result to variable `country_data_sorted_by_pop`:

````python
country_data_sorted_by_pop = country_data.sort_values(by='population', ascending=False).reset_index()
country_data_sorted_by_pop
````

## Computing values from a `DataFrame`

You can apply **arithmetic operations** on entire data frames or series,
such as selected columns.

Let's calculate the _population density_ (_population_ divided by _area_) for each country:

````python
country_data['population'] / country_data['area']
````

### Adding a new column to a `DataFrame`

It is more useful to combine the computed population density with the existing data,
by **adding a new column** with the computed density values.  
We put it in a new column labeled `'density'` as follows:

````python
country_data['density'] = country_data['population'] / country_data['area']
country_data
````

### Boolean masks

Such computations can also be used to calculate boolean values.
For instance, let's compute which countries are in _Europe_.  
We assign it to variable `countries_in_EU` for later use.

````python
countries_in_EU = country_data['continent'] == 'EU'
countries_in_EU
````

This may not seem very useful.
But wait and see.

Such a column of boolean values is also called a **boolean mask**.
It can be used to select rows,
as explained next.

## Selecting rows by boolean masks

A **boolean mask** can be used to select those rows
for which the mask contains the value `True`.
This is done by using the boolean mask as index of the data frame.

So, let's find those countries in _Europe_:

````python
country_data[countries_in_EU]
````

And here is the data for the _Netherlands_,
knowing that is has _top-level domain_ `.nl`:

````python
country_data[country_data['tld'] == '.nl']
````

### Boolean mask for missing values

The `Series` function **`isnull()`** can be used to create a **boolean mask
for missing values** (these are also known as **not available** (na) or **null values**).

Which countries don't have a _top-level domain_:

````python
country_data[country_data['tld'].isnull()]
````

### Combining boolean masks

Boolean masks can be combined into more complex boolean conditions using the following operators:

* **`&`** for 'and'
* **`|`** for 'or'
* **`~`** for 'not'

These bind stronger than comparison operators
(use parenthesis to evaluate comparisons first; see next example).

Let's show data for contries with a _population_ less than one million and
an _area_ more than one hundred thousand square kilometer:

````python
country_data[(country_data['population'] < 1_000_000) & (country_data['area'] > 100_000)]
````

## Grouping rows in a `DataFrame`

By sorting a data frame,
you can bring items with the same value together.
But this is not good enough to gather statistics
about groups of equal values.

Take for instance the country data.
What about total _population_ and _area_ per _continent_?

That is where **grouping** comes in.
First, we create a so-called `GroupBy` object
from the data frame,
using the function **`groupby()`**
with the column label(s) to group by as argument.

This is how you group the country data by _continent_,
and assign the `GroupBy` object to variable `grouped`.

````python
grouped = country_data.groupby('continent')
````

The **`groups`** attribute of the `GroupBy` object
shows for each continent which rows concern that continent:

````python
grouped.groups
````

> Note that Australia and New Zealand are apparently put under Oceania,
> since there is no separate continent `AU`.

The `GroupBy` object can be treated like a `DataFrame` in a number of ways.

For instance, this is how you get the summary statistics for _area_ per _continent_:

````python
grouped['area'].describe()
````

And this is how you get total _population_ and _area_ per _continent_:

````python
grouped[['population', 'area']].sum()
````

## Modifying a `DataFrame`

`DataFrame` objects are _mutable_.
In this lesson we will only present two very basic ways of modifying a `DataFrame` object.

### Changing a value in a `DataFrame`

To change a value at a particular location of `DataFrame df`,
you can use an assignment to the selected item:
```python
df.loc[row, column] = expression
```

````python
df.loc[0, 'B'] = 'ACE'
df
````

### Deleting rows or columns from a `DataFrame`

To delete one or more rows or columns from `DataFrame df`
use the data frame function **`df.drop()`**.

* **`df.drop(row)`**
* **`df.drop(list_of_rows)`**
* **`df.drop(column, axis=1)`**
* **`df.drop(list_columns, axis=1)`**

This returns a _new data frame_, unless you supply the extra argument **`inplace=True`**.  
Note: This is usually not recommended; it is often better to assign the result to a new variable.

The parameter `axis` indicates along which axis to drop.
The default is `axis=0`, to drop rows.

You can also **drop rows based on a condition** using `drop()`,
by cleverly combining

* selection with a boolean mask and
* the `index` attribute:

```python
df.drop(df[boolean_mask].index]
```

````python
df.drop(1)  # drop the row with index label 1 (the second row)
````

````python
df.drop('C', axis=1)  # drop the column with label 'C'
````

````python
df.drop([1, 3], inplace=True)  # drop rows with index labels 1 and 3, modifying df in place
df
````

````python
country_data.drop(country_data[countries_in_EU].index)  # drop countries in Europe
````

## Plotting a `DataFrame`

Various plots can be made directly from a `DataFrame` or `Series` object.

First, we

* import the underlying plotting libraries **Matplotlib** and **Seaborn**,
* issue the command **`%matplotlib inline`**
    to make plots appear inside the notebook,
* and configure some parameters (see the comments for details).

````python
# next command ensures that plots appear inside the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  # set Seaborn defaults
plt.rcParams['figure.figsize'] = 10, 5  # default hor./vert. size of plots, in inches
plt.rcParams['lines.markeredgewidth'] = 1  # to fix issue with seaborn box plots; needed after import seaborn
````

> NOTE: The plotting methods that we present here are for **personal plots**.  
> You know the data you are plotting, so you can interpret these plots
> but someone else not.    
> For **communication to others**, plots need to include more information,
> and need to adhere to more stringent standards.
>
> Furthermore, this lesson does not cover all relevant plot types.
>
> These things are addressed in _Programming Lesson and Exercises - Matplotlib and Seaborn_.

The general function to make a plot for a `DataFrame df` is

* **`df.plot(kind='...', ...)`**

where the following kinds of plots are supported.

* **`'line'`** : default
* **`'area'`** : like `'line'`, but with area below line filled in
* **`'bar'`** : vertical bar plot, with separated bars
* **`'barh'`** : horizontal bar plot, with separated bars
* **`'hist'`** : histogram, with automatic binning; use `bins=n` to set the number of bins to `n`
* **`'box'`** : vertical box (and whisker) plot, with outliers
* **`'density'`** : density plot (a kind of continuous histogram)
* **`'scatter'`** : scatter plot

All of these plots, **except `'scatter`'**, apply to `Series` objects as well.

When used on a `DataFrame` object, all (numeric) columns are plotted,
and a **legend** is included.

The extra arguments are mostly irrelevant in this lesson;
the ones that are relavant will be discussed below.

For **box plots** of `DataFrame` objects, there is a special variant:

* **`df.boxplot(column=..., by=...)`** :
    side-by-side box plots for given column(s), grouped by given column(s)

In this lesson, we will stick to default plotting behavior,
which is good enough for _personal plots_.

Remember that you can get built-in help with `?`.  
Execute the following cell to get built-in help for the Pandas plotting function
(most of the details are irrelevant for this lesson):

````python
df.plot?
````
