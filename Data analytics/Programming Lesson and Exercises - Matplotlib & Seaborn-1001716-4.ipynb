## Concepts and terminology for plotting

Although we have tried to mininize the need to use `matplotlib` directly,
you do need to know a few things about `matplotlib`.

* `matplotlib` uses the same approach to plotting as _MatLab_
    (a programming language you may already have seen, or will see in other technical courses).
* With `matplotlib` you **build up a plot in a sequence of steps**,
    where each step adds _plot elements_ to the _**current figure**_.
* In a Jupyter notebook (like this one),
    such a plot must be built up in a **single code cell**.
* When that code cell is executed,
    * the _current figure_ is **automatically cleared and initialized** at the beginning, and
    * the resulting plot is **automatically shown** at the end.
* Therefore, you **do _not_ call `plt.figure()` and `plt.show()`** in a Jupyter notebook.
    (You will often see these commands when browsing the web for solutions.
    Don't use them.)


The image on the right shows some standard terminology for plot elements.

* **Figure**: refers to the whole figure,
    possibly consisting of multiple plots ('_axes_') arranged in a grid
* **Axes** (plural): an actual plot visualizing data in its central area,
    bounded by _spines_ that show _axis_ information;
    not to be confused with just the $x$- and $y$-axis
* **Axis** (singular): information that defines the relationship between data values and
    horizontal or vertical offsets in the plot
* **Spine**: a line on the boundary with reference information obtained from an _axis_
* **Major and minor ticks**: longer and shorter lines marking off positions on a _spine_
* **Major and minor tick labels**: numeric or textual labels of _ticks_
* **Axis label**: a textual label of an _axis_
* **Title**: textual label of a plot
* **Grid**: auxiliary reference lines derived from each _axis_
* **Line** and **markers**: visualizations of data values
* **Legend**: overview that associates text with visual elements for data

Also see the `matplotlib` [Usage Guide](https://matplotlib.org/tutorials/introductory/usage.html).

## Title, axis labels, legend

To add elements as a **title**, **axis labels**, and a **legend**,
we need to capture the result of the plot call in a variable.
It is customary to name this variable `ax` or `axes`.

This variable can then be used as follows to add certain plot elements.

* **`ax.set_title(string, ...)`** : set the plot's title
* **`ax.set_xlabel(string, ...)`** : set the $x$-axis label
* **`ax.set_ylabel(string, ...)`** : set the $y$-axis label
* **`ax.legend(list_of_strings)`** : add a legend, one string per column plotted

The functions `ax.set_...` can take additional keyword arguments:

* **`size=...`** (a number, font size in points)
* **`weight=...`** (a string, font weight, e.g. `'bold'`)

For details see: [Text properties and layout](https://matplotlib.org/users/text_props.html).

We create the same plot, but now with additional information,
using the variable `ax_1x` (so that we can reuse that later):

````python
ax_1x = aq1['x'].plot()
ax_1x.set_title("Anscombe's Quartet", size=18, weight='bold')
ax_1x.set_title("Data Set I", loc='right', size=18, weight='bold')
ax_1x.set_xlabel('Index')
ax_1x.set_ylabel("value in column 'x'")
ax_1x.legend(['x']);
````

## Plot range limits and plot size

To illustrate some additional plotting elements,
we introduce another data set named **`df_grade`**.
It consists of grades for 2400 students,
drawn from a _normal distribution_ with mean 7 and standard deviation 1,
clipped to the interval from 0 through 10, rounded to one decimal.  
(Execute the following code cell.)

````python
df_grade = pd.DataFrame(np.random.normal(7, 1, 2400), columns=['Grade'])
df_grade = np.clip(df_grade, 0, 10)  # reduce to interval [0, 10]
df_grade = np.round(df_grade, decimals=1)  # round to one decimal
df_grade.describe()
````

First, we show a density plot without tweaking it:

````python
ax = df_grade.plot(kind='density', legend=False);
````

You could find it annoying that the grade scale on the $x$-axis extends beyond grade 10.

There is a way to **limit the plot range** by using

* **`ax.set_xlim(min, max)`**
* **`ax.set_ylim(min, max)`**

````python
ax = df_grade.plot(kind='density', legend=False)
ax.set_xlim(0, 10);  # limit the x-range
````

The **size of a plot** can be set for a single plot
by supplying the keyword argument **`figsize=(W, H)`**,
where `W` and `H` are the width and height respectively, in _inches_.

### Juxtaposition (subplots)

Multiple plots can also be combined by **juxtapostion** (placing them **next to each other**)
**in a single figure**,
rather then superimposing them.

One way of doing that for data from a **single data frame**
that is plotted via a **single `plot()` call**
is via the argument **`subplots=True`**.  
Compare this to the default superimposed plot above.

````python
aq1.plot(subplots=True);
````

Putting plots that are created by **separate `plot()` calls**
together **in a single figure**
is done via the _axes_ object.

There are various ways to do this.
We present a uniform way,
which works both with Pandas and Seaborn.

First, you set up a **grid of subplots** using plain `matplotlib` code.
The common idiom for this is:
```python
fig, ax = plt.subplots(nrows=..., ncols=..., squeeze=False, ...)
```

> **NOTE**:
> The argment **`squeeze=False`** ensures that `ax` will always be a 2D array of axes objects.  
> With `squeeze=True` (which is the default),
> if one of the dimensions equals 1,
> `ax` will be a 1D array, representing a row (if `nrows=1`) or a column (if `ncols=1`).  
> We recommend `squeeze=False`,
> so that you can always write `ax[i, j]` to select the appropriate subplot.

Let's try that for a 2x2 grid with _empty_ plots.

````python
fig, ax = plt.subplots(nrows=2, ncols=2, squeeze=False)
````

You can forget about the variable `fig`;
we don't need that, but it is needed to capture a value returned by `plt.subplots()`.

The variable `ax` now holds a **two-dimensional array of _axes_ objects**.  
It can be indexed by **`ax[row][column]`** or **`ax[row, column]`**,
where **indexing starts with 0**.

Next, you provide each plot call with the appropriate argument **`ax=ax[row, column]`**.

Here is all the data in Anscombe's Quartet,
shown in four separate plots,
plotting the `'x'` and `'y'` values against their index.

````python
fig, ax = plt.subplots(nrows=2, ncols=2, squeeze=False)
aq1.plot(style=':o', ax=ax[0, 0])
aq2.plot(style=':o', ax=ax[0, 1])
aq3.plot(style=':o', ax=ax[1, 0])
aq4.plot(style=':o', ax=ax[1, 1]);
````

You can supply extra arguments to `plt.subplots()`:

* **`sharex=True`** shares the $x$-axis
* **`sharey=True`** shares the $y$-axis

You can set a **global title** with `fig.suptitle()`.

The preceding plots, but now with shared $x$- and $y$-axis,
and with both subplot titles and a global title:

````python
fig, ax = plt.subplots(nrows=2, ncols=2, squeeze=False, sharex=True, sharey=True)
aq1.plot(style=':o', ax=ax[0, 0])
aq2.plot(style=':o', ax=ax[0, 1])
aq3.plot(style=':o', ax=ax[1, 0])
aq4.plot(style=':o', ax=ax[1, 1])
ax[0, 0].set_title('Data Set I')
ax[0, 1].set_title('Data Set II')
ax[1, 0].set_title('Data Set III')
ax[1, 1].set_title('Data Set IV')
fig.suptitle("Line plots for Anscombe's Quartet", size=18, weight='bold');
````

## Scatter plots: variations

### Scatter matrix

When investigating data with multiple variables and/or many values,
it is important to take a bit more care when creating plots.

We illustrate this with some examples based on another (generated) data set.
(It does not matter how this data is generated; you can take that for granted.)

The data has _four_ variables, named `a`, `b`, `c`, and `e`,
where `a`, `b`, and `e` are _numeric_, and `c` (for condition) is _boolean_:

````python
df_sd = pd.DataFrame(((x, v, c, e)
                      for c in (False, True)
                      for x in np.random.sample(5000 if c else 20000)
                      for e in np.random.sample(1)
                      for v in [x**3 + (e-0.5)/10 if c else x + (e - 0.8)*0.7]
                      if 0 <= v < 1),
                     columns=list('abce'))
df_sd.info()
````

````python
df_sd.head()
````

To get an impression of possible relationships between these variables,
we create a **scatter matrix**,
consisting of a **scatter plot** for _each pair_ of the numeric variables
(`a`, `b`, `e`).
On the diagonal of this matrix we show a **density plot** of the corresponding variable.

This is accomplished by **`pd.plotting.scatter_matrix()`**,
supplying as arguments the appropriate columns of the data frame,
and the keyword argument **`diagonal='...'`**:

````python
# scatter matrix for pairs of colums 'a', 'b', and 'e'
pd.plotting.scatter_matrix(df_sd[list('abe')], diagonal='density', figsize=(10,10));  # all pairwise scatter plots; density plot on diagonal
````

> **NOTE**: Each scatter plot appears twice, with the main diagonal at their center.  
> The two axes are swapped (the plot is _transposed_).

The variables `a` and `b` appear to be _positively correlated_:
higher values of `a` correspond to higher values of `b`.

There are also peculiar patterns in the other two correlations.

However, because there are so many values in the plots,
details may have been obscured.
There is _lack of contrast_.

### Transparency (alpha setting), point size

There are two parameters that we can play with to improve the contrast:

* the **transparency** of the dots (also knowns as the **alpha setting**), and
* the **size** of the dots.

The **alpha setting** is a value between

* 0 (fully _transparent_) and
* 1 (fully _opaque_).

When dots are plotted on top of each other in an area,
their opaqueness accumulates: more dots means more ink.

We illustrate this with the scatter plot for `a` and `b`,
using two different alpha setings by supplying **`alpha=`** as extra argument:

````python
# effect of alpha (transparency: 0 = transparent; 1 = opaque)
fig, ax = plt.subplots(ncols=2, sharey=True)
df_sd.plot(kind='scatter', x='a', y='b', alpha=1, ax=ax[0])  # too high
df_sd.plot(kind='scatter', x='a', y='b', alpha=0.005, ax=ax[1]);  # a bit low
````

### Scatter plot with regression curve

Seaborn provides the function **`reg_plot()`** to create a **scatter plot**
with a **linear regression curve** and the **95% confidence interval** for that curve.
By default, a first-order linear regression is done;
that is, on the plain data, without higher-order terms.

> **NOTE**:
> The concept of _linear regression_ will be treated in more detail in the lesson on _SciKit Learn_.
>
> Briefly stated, the _linear regression curve_ for a collection of data points (as in a scatter plot)
> is the curve that _best fits_ the data.
> For an _order-$n$_ linear regression, the curve is a _polynomial_ of degree (at most) $n$.
> For $n=1$, this is a straight line.
>
> The more the points are scattered, the looser the fit,
> resulting in a wider area for the 95% confidence interval.

Here is an example for data set I in Anscombe's Quartet,

````python
sns.regplot(data=aq1, x='x', y='y');
````
